Task1 : write a Python function that reads
a list of numbers and calculates the mean, minimum, and
maximum values.
Prompt : write a Python function that reads a list of numbers and calculates the mean, minimum, and maximum values and enter number from the console.


```python
def analyze_numbers():
  """
  Reads a list of numbers from the console and calculates the mean, minimum, and maximum values.
  """
  numbers = []
  while True:
    try:
      num_str = input("Enter a number (or press Enter to finish): ")
      if not num_str:
        break
      number = float(num_str)
      numbers.append(number)
    except ValueError:
      print("Invalid input. Please enter a number.")

  if not numbers:
    print("No numbers entered.")
    return

  mean = sum(numbers) / len(numbers)
  minimum = min(numbers)
  maximum = max(numbers)

  print(f"Numbers entered: {numbers}")
  print(f"Mean: {mean}")
  print(f"Minimum: {minimum}")
  print(f"Maximum: {maximum}")

analyze_numbers()
```

This Python code defines a function called analyze_numbers() that does the following:

1 . Initializes an empty list: It creates an empty list called numbers to store the numbers entered by the user.
2 . Reads input from the console: It enters a while loop that continues indefinitely until the user decides to stop. Inside the loop, it prompts the user to "Enter a number (or press Enter to finish):" using the input() function.
3 . Handles user input:
> If the user just presses Enter without typing anything (if not num_str:), the loop breaks, indicating the user is finished entering numbers.
> If the user enters something, it tries to convert the input string (num_str) into a floating-point number using float(num_str).
> If the conversion is successful, the number is added to the numbers list using numbers.append(number).
> If the conversion fails (e.g., the user enters text that isn't a number), a ValueError is caught, and an error message "Invalid input. Please enter a number." is printed.
4 . Checks if any numbers were entered: After the loop finishes, it checks if the numbers list is empty (if not numbers:). If it is, it prints "No numbers entered." and the function ends.
5 . Calculates statistics: If numbers were entered, it calculates:
The mean (average) by summing all the numbers in the list (sum(numbers)) and dividing by the total count of numbers (len(numbers)).
The minimum value using the min() function.
The maximum value using the max() function.
6 . Prints the results: Finally, it prints the list of numbers entered, the calculated mean, the minimum value, and the maximum value using f-strings for formatted output.
7 . Calls the function: The last line analyze_numbers() calls the function to start the process of reading numbers and performing the analysis.

```python
def is_armstrong_number():
    """
    Checks if a number entered from the console is an Armstrong number.
    """
    num_str = input("Enter an integer: ")
    try:
        num = int(num_str)
    except ValueError:
        print("Invalid input. Please enter an integer.")
        return

    # Calculate the number of digits
    num_digits = len(num_str)

    # Calculate the sum of the cubes of the digits
    sum_of_powers = 0
    temp_num = num
    while temp_num > 0:
        digit = temp_num % 10
        sum_of_powers += digit ** num_digits
        temp_num //= 10

    # Check if it's an Armstrong number
    if num == sum_of_powers:
        print(f"{num} is an Armstrong number.")
    else:
        print(f"{num} is not an Armstrong number.")

is_armstrong_number()
```

```python
def is_prime():
    """
    Checks if a number entered from the console is a prime number.
    """
    num_str = input("Enter an integer: ")
    try:
        num = int(num_str)
    except ValueError:
        print("Invalid input. Please enter an integer.")
        return

    if num <= 1:
        print(f"{num} is not a prime number.")
        return

    for i in range(2, int(num**0.5) + 1):
        if num % i == 0:
            print(f"{num} is not a prime number.")
            return

    print(f"{num} is a prime number.")

is_prime()
```

Tak 5 : Students need to write a Python program to calculate the sum of
odd numbers and even numbers in a given tuple.

odd numbers and even numbers in a given tuple.and take number number from console

```python
def sum_odd_even_in_tuple():
    """
    Calculates the sum of odd and even numbers in a tuple created from user input.
    """
    num_str_list = input("Enter numbers separated by spaces: ").split()
    numbers = []
    for num_str in num_str_list:
        try:
            numbers.append(int(num_str))
        except ValueError:
            print(f"Skipping invalid input: {num_str}")

    if not numbers:
        print("No valid numbers entered.")
        return

    num_tuple = tuple(numbers)
    print(f"Created tuple: {num_tuple}")

    # Proceed to calculate sums in the next step
    odd_sum = 0
    even_sum = 0
    for number in num_tuple:
        if number % 2 == 0:
            even_sum += number
        else:
            odd_sum += number

    print(f"Sum of odd numbers: {odd_sum}")
    print(f"Sum of even numbers: {even_sum}")

sum_odd_even_in_tuple()
```
